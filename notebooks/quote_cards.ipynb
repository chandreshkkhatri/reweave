# Quote Cards Workflow

Generate beautiful inspirational quote cards with:
- 1080x1080 square image (Instagram feed)
- 1080x1920 vertical image (Instagram story)
- 15-second Ken Burns zoom video with TTS narration (TikTok/Reels)

## Prerequisites
- Set `GEMINI_API_KEY` in your `.env` file
- Install dependencies: `pip install reweave`

```python
from reweave.workflows.quote_card_workflow import QuoteCardWorkflow

workflow = QuoteCardWorkflow()
```

```python
# Generate a stoic wisdom quote card
result = workflow.generate_quote_card(
    content_id="stoic-001",
    theme="stoic wisdom"
)

print(f"Quote: {result['quote']['quote_text']}")
print(f"By: {result['quote']['attribution']}")
print(f"\nOutputs:")
print(f"  Square image: {result['square_image']}")
print(f"  Story image: {result['story_image']}")
print(f"  Video: {result['video']}")
```

```python
# Generate a Monday motivation quote card
result = workflow.generate_quote_card(
    content_id="motivation-001",
    theme="monday motivation"
)
```

```python
# Batch generate multiple themes
themes = ["perseverance", "leadership", "creativity", "gratitude"]

for i, theme in enumerate(themes):
    print(f"\n--- Generating: {theme} ---")
    result = workflow.generate_quote_card(
        content_id=f"{theme}-001",
        theme=theme
    )
    print(f"Done: {result['quote']['quote_text']}")
```
