# Podcast Highlight Clips Workflow

Extract the best moments from a long podcast and generate individual short-form clips:
- Fetches transcript from YouTube
- Gemini identifies the N most clip-worthy segments
- Generates vertical video (1080x1920) for each clip with title card + scrolling text + TTS

## Prerequisites
- Set `GEMINI_API_KEY` in your `.env` file
- Install dependencies: `pip install reweave`

```python
from reweave.workflows.podcast_clips_workflow import PodcastClipsWorkflow

workflow = PodcastClipsWorkflow()
```

```python
# Generate 5 highlight clips from a podcast
result = workflow.generate_clips(
    content_id="podcast-001",
    youtube_url="https://www.youtube.com/watch?v=YOUR_VIDEO_ID",
    num_clips=5
)

print(f"Podcast: {result['podcast_title']}")
print(f"Generated {len(result['clips'])} clips:")
for clip in result['clips']:
    print(f"  {clip['number']}. {clip['title']}")
    print(f"     Video: {clip['video']}")
```

```python
# Generate fewer clips for a shorter video
result = workflow.generate_clips(
    content_id="podcast-002",
    youtube_url="https://www.youtube.com/watch?v=YOUR_VIDEO_ID",
    num_clips=3
)
```
