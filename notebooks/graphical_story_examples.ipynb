# Graphical Story Workflow – Example Inputs

This notebook demonstrates a variety of story prompts for `GraphicalStoryWorkflow`.
Each example shows how `content_id`, `title`, and `additional_instructions` shape the output.

**Prerequisites:** Set `GEMINI_API_KEY` in your `.env` file.

```python
from reweave.workflows.graphical_story_workflow import GraphicalStoryWorkflow

workflow = GraphicalStoryWorkflow()
```

---
## Example 1 – Classic Fable

A timeless fable with a moral lesson, illustrated in a warm watercolour style.

```python
content_id   = "tortoise_hare_1"
title        = "The Tortoise and the Hare"
instructions = (
    "Set the story in a lush forest with anthropomorphic animals. "
    "Use a warm watercolour illustration style. "
    "Emphasise the moral: slow and steady wins the race."
)
```

```python
workflow.create_story(content_id, title, instructions)
workflow.generate_script(content_id)
```

```python
workflow.generate_footages(content_id)
```

```python
workflow.generate_final_video(content_id)
```
